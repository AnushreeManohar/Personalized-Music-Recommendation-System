## Load the dataset

```python
df1 = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/ralmuta@gmu.edu/final_data_df.csv")
df1.head(5)
```

```python
display(df1)
```

#### Queries

##### 1. What was the top songs in the 90s?

```python
df_sel = df1.filter((df1.popularity>=1) & (df1.year>= 1990) & (df1.year<= 1999) ).select(df1.name, df1.artists, df1.year)
display(df_sel)
```

##### 2. What was the most popular songs in FL state between 1980 and 1990?

```python
df_sel = df1.filter((df1.popularity>=1) & (df1.state == 'FL') & (df1.year>= 1980) & (df1.year<=1990)).select(df1.name, df1.year, df1.state, df1.artists)
display(df_sel)
```

##### 3. Who was the most popular female artists in the 90s?

```python
df_sel = df1.filter((df1.popularity>=1) & (df1.Gender == 'F') & (df1.year>= 1990) & (df1.year<= 1999)).select(df1.name, df1.artists, df1.Gender, df1.year)
display(df_sel)
```

##### 4. When listening period is 4, How many people are listening to songs that acousticness is less than 0.5?

```python
display(df1.select("User_id").where((df1.acousticness<0.5) & (df1.Listening_Period == 4)).count())
```

##### 5. What songs has more loudness and more danceability?

```python
df_sel = df1.filter((df1.loudness>-20) & (df1.danceability>= 0.5)).select(df1.name, df1.artists, df1.danceability, df1.loudness) 
df_sel.display()
```

##### 6. What is the average time people from each state listen to music?

```python
from pyspark.sql.functions import avg, round
df_sel = df1.groupBy('state').agg(round(avg('Listening_Period'), 1).alias('avg_Listening_Period'))
display(df_sel)
```

##### 7. What are the top songs with high energy and danceability?

```python
df_sel = df1.filter((df1.danceability>=0.7) & (df1.energy>=0.7) & (df1.popularity>=1)).select(df1.name, df1.artists, df1.danceability, df1.energy)
display(df_sel.sort(df1.duration_ms.cast('int').asc()))
```
